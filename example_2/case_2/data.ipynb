```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt


import data
```

```python
x, f = data.generate_data_2(l=0.1, M=256, N=10000, K=2)

ws = np.array([0.5] + 255 * [1] + [0.5]).reshape([-1, 1])
k = (np.sum(f**2 * ws, axis=0) / 256)

sio.savemat(
    "KL_2.mat", {"t": x, "f": f, "k": k},
)

x, f = data.generate_data_2(l=0.1, M=256, N=10000, K=4)

ws = np.array([0.5] + 255 * [1] + [0.5]).reshape([-1, 1])
k = np.sqrt(np.sum(f**2 * ws, axis=0) / 256)

sio.savemat(
    "KL_4.mat", {"t": x, "f": f, "k": k},
)

x, f = data.generate_data_2(l=0.1, M=256, N=10000, K=6)

ws = np.array([0.5] + 255 * [1] + [0.5]).reshape([-1, 1])
k = np.sqrt(np.sum(f**2 * ws, axis=0) / 256)

sio.savemat(
    "KL_6.mat", {"t": x, "f": f, "k": k},
)

x, f = data.generate_data_2(l=0.1, M=256, N=10000, K=8)

ws = np.array([0.5] + 255 * [1] + [0.5]).reshape([-1, 1])
k = (np.sum(f**2 * ws, axis=0) / 256)

sio.savemat(
    "KL_8.mat", {"t": x, "f": f, "k": k},
)

x, f = data.generate_data_2(l=0.1, M=256, N=10000, K=12)

ws = np.array([0.5] + 255 * [1] + [0.5]).reshape([-1, 1])
k = (np.sum(f**2 * ws, axis=0) / 256)

sio.savemat(
    "KL_12.mat", {"t": x, "f": f, "k": k},
)

x, f = data.generate_data_2(l=0.1, M=256, N=10000, K=16)

ws = np.array([0.5] + 255 * [1] + [0.5]).reshape([-1, 1])
k = (np.sum(f**2 * ws, axis=0) / 256)

sio.savemat(
    "KL_16.mat", {"t": x, "f": f, "k": k},
)
```
