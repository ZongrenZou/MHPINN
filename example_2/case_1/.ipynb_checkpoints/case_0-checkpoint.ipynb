```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import tensorflow as tf
import time

import models
import flows
```

```python
data = sio.loadmat("data_10.mat")
t = data["t"]
u_ref = data["sols"]
f = data["f"]
l = data["l"]
```

```python
t_train = t[::8]
f_train = f[::8, 0:1]
u_exact = u_ref[:, :, 0]
```

```python
model = models.PINN(eps=0)
```

```python
loss = model.train(t_train, f_train, niter=20000)
```

```python
model.restore()
u_pred = model.call(
    tf.constant(t, tf.float32)
).numpy()
f_pred = model.pde(
    tf.constant(t, tf.float32)
).numpy()
```
