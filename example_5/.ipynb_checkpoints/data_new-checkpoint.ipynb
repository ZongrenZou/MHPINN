```python
x = np.linspace(0, 2*np.pi, 101)
y = np.linspace(0, 2*np.pi, 101)
xx, yy = np.meshgrid(x, y)

N = 200
D = 25
xi = np.random.uniform(size=[N, D*4])
# xi = np.random.normal(size=[N, D*4])
f = np.zeros([N, 101, 101])

for i in range(D):
    print(4*i, 4*i+1, 4*i+2, 4*i+3)
    print(i+1)
    f = f + xi[:, 4*i:4*i+1][..., None] * np.sin((i+1)/5*xx)[None, ...] \
        + xi[:, 4*i+1:4*i+2][..., None] * np.cos((i+1)/5*xx)[None, ...] \
        + xi[:, 4*i+2:4*i+3][..., None] * np.sin((i+1)/5*yy)[None, ...] \
        + xi[:, 4*i+3:4*i+4][..., None] * np.cos((i+1)/5*yy)[None, ...]
# for i in range(D):
#     print(4*i+2, 4*i+3)
#     f = f + xi[:, 4*i+2:4*i+3][..., None] * np.sin((i+1)*yy)[None, ...] \
#         + xi[:, 4*i+3:4*i+4][..., None] * np.cos((i+1)*yy)[None, ...]

f = 2/D/4 * f 


# # for i in range(D):
# #     f = f + xi[:, 4*i:4*i+1][..., None] * np.sin((i+1)*xx)[None, ...] \
# #         + xi[:, 4*i+1:4*i+2][..., None] * np.cos((i+1)*xx)[None, ...] \
# #         + xi[:, 4*i+2:4*i+3][..., None] * np.sin((i+1)*yy)[None, ...] \
# #         + xi[:, 4*i+3:4*i+4][..., None] * np.cos((i+1)*yy)[None, ...]
# # f = 2/D * f
```

```python
sio.savemat(
    "source_100.mat",
    {
        "f": f, "x": x, "y": y, "xx": xx, "yy": yy,
        "xi": xi
    }
)
```
