## Case 1-2

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import tensorflow as tf
# import seaborn as sns
import time

import models
import flows
```

```python
data = sio.loadmat("case3.mat")
x1 = data["x1"]
x2 = data["x2"]
y1 = data["y1"]
y2 = data["y2"]
w1 = data["w1"]
w2 = data["w2"]
A = data["A"]
k = data["k"]

x1_train = x1[0, :].reshape([-1, 1])
y1_train = y1[:500].T
x2_train = x2[0, :].reshape([-1, 1])
y2_train = y2[:500].T
w1_train = w1[:500]
w2_train = w2[:500]
A1_train = A[:500]
A2_train = A[2000:2500]
k1_train = k[:500]
k2_train = k[2000:2500]
```

```python
meta = models.Meta(num_tasks=1000, dim=50)
```

```python
loss = meta.train(
    x1_train,
    y1_train,
    x2_train,
    y2_train,
    niter=50000,
    ftol=1e-10,
)
```

```python
meta.restore()
loss[-1], min(loss)
```

#### Distribution learning

```python
permutation = list(np.arange(26, 51, 1)) + list(np.arange(0, 26, 1))
nf = flows.MAF(
    dim=51,
    permutation=permutation,
    hidden_layers=[100, 100],
    num_bijectors=10,
    activation=tf.nn.relu,
    name="maf",
)
```

```python
loss = nf.train_batch(tf.constant(meta.heads.numpy().T, tf.float32), nepoch=1000)
```

```python
nf.restore()
heads = nf.sample(1000)
x_test = tf.constant(np.linspace(-1, 1, 100).reshape([-1, 1]), tf.float32)
# y_test = tf.constant(y_train, tf.float32)

out = meta.call(x_test, tf.transpose(heads))
for i in range(1000):
    plt.plot(x_test, out[:, i])
plt.ylim([-6, 6])

plt.figure()
out = meta.call(x_test, meta.heads)
for i in range(1000):
    plt.plot(x_test, out[:, i])
plt.ylim([-6, 6])
plt.show()
```

```python
nf.restore()
heads = nf.sample(1000).numpy()
x_test = tf.constant(np.linspace(-1, 1, 100).reshape([-1, 1]), tf.float32)

out = meta.call(x_test, tf.transpose(heads))
fig = plt.figure(dpi=200)
ax = fig.add_subplot()
for i in range(1000):
    ax.plot(x_test, out[:, i], linestyle="-")
ax.set_box_aspect(2/3)
ax.set_ylim([-6, 6])
fig.savefig("samples_learned_distribution.png")

out = meta.call(x_test, meta.heads)
fig = plt.figure(dpi=200)
ax = fig.add_subplot()
for i in range(500):
    ax.plot(x_test, A1_train[i] * np.cos(w1_train[i]*x_test) + 2*k1_train[i]*x_test, linestyle="-")
for i in range(500):
    ax.plot(x_test, A2_train[i] * np.cos(w2_train[i]*x_test) + 2*k2_train[i]*x_test, linestyle="-")
ax.set_box_aspect(2/3)
ax.set_ylim([-6, 6])
fig.savefig("samples_prior_distribution.png")
```
