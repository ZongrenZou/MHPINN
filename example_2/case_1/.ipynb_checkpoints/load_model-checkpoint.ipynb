```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import tensorflow as tf
import time

import models
import flows
```

```python
data = sio.loadmat("./data/data_10.mat")
t = data["t"]
u_ref = data["sols"]
f = data["f"]
l = data["l"]

t_train = t[::4]
f_train = f[::4, :2000]
u_ref = u_ref[..., :2000]
```

```python
meta = models.Meta(num_tasks=2000, dim=50, name="meta_10_4")
meta.restore()
```

```python
t_test = np.linspace(0, 1, 257).reshape([-1, 1])
u_pred = meta.call(
    tf.constant(t_test, tf.float32), meta.heads
).numpy()
i = 2
plt.plot(t_test, u_pred[:, i], '--')
plt.plot(t_test, u_ref[:, 0, i])
plt.show()
```

```python
L2 = np.sqrt(np.sum((u_pred - u_ref[:, 0, :])**2, axis=0) / np.sum(u_ref[:, 0, :]**2, axis=0))
```

```python
print(np.mean(L2))
```

```python
permutation = list(np.arange(26, 51, 1)) + list(np.arange(0, 26, 1))

nf = flows.MAF(
    dim=51,
    permutation=permutation,
    hidden_layers=[128, 128],
    num_bijectors=10,
    activation=tf.nn.relu,
    name="flow_10_4",
    
)
nf.restore()
```

```python
sample_fn = tf.function(nf.sample)
heads = sample_fn(1000)
t_test = tf.constant(np.linspace(0, 1, 257).reshape([-1, 1]), tf.float32)

u_pred = meta.call(t_test, tf.transpose(heads))
for i in range(1000):
    plt.plot(t_test, u_pred[:, i])
plt.ylim([-1.0, 1])
    
plt.figure()
for i in range(1000):
    plt.plot(t_test, u_ref[:, 0, i])
plt.ylim([-1.0, 1])
```

```python
heads = sample_fn(2000)
t_test = tf.constant(np.linspace(0, 1, 257).reshape([-1, 1]), tf.float32)
f_pred = meta.pde(t_test, tf.transpose(heads))
fig = plt.figure(dpi=100)
ax = fig.add_subplot()
for i in range(2000):
    ax.plot(t_test, f_pred[:, i], linestyle="-")
ax.set_ylim([-4, 4])
# ax.set_title("Learned")
# fig.savefig("case_1/case_1_5_learned.png")
# plt.savefig("case_1/learned_distribution.png")
```

```python
f_ref = meta.pde(t_test, meta.heads)
fig = plt.figure(dpi=100)
ax = fig.add_subplot()
for i in range(2000):
    ax.plot(t_test, f[:, i], linestyle="-")
ax.set_ylim([-4, 4])
# ax.set_title("Target")
#fig.savefig("case_1/case_1_5_target.png")
# plt.savefig("case_1/target_distribution.png")
```

```python
import tensorflow_probability as tfp

K = tfp.stats.covariance(tf.transpose(f_pred)) / 256
eigs = tf.linalg.eigvals(K)
eigs = tf.cast(eigs, tf.float32)

K = tfp.stats.covariance(tf.transpose(f[:, :2000])) / 256
eigs2 = tf.linalg.eigvals(K)
eigs2 = tf.cast(eigs2, tf.float32)

K = tfp.stats.covariance(tf.transpose(f)) / 256
eigs3 = tf.linalg.eigvals(K)
eigs3 = tf.cast(eigs3, tf.float32)
```

```python
plt.plot(eigs[::-1][:20], 'o-', label="learned")
plt.plot(eigs2[::-1][:20], 's-', label="training")
plt.plot(eigs3[::-1][:20], '^-', label="referece")
plt.legend()
plt.title("Top 20 eigenvalues")
```
