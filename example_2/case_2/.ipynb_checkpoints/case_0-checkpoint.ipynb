```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import tensorflow as tf
import time

import models
import flows
```

```python
data = sio.loadmat("./dataset/data_8.mat")
t = data["t"]
u_ref = data["sols"]
f = data["f"]
k = data["k"]
```

```python
t.shape, u_ref.shape, f.shape, k.shape
```

```python
j = 193
t_f_train = t[::8]
f_train = f[::8, j:j+1]
t_u_train = t[::32]
u_train = u_ref[::32, 0, j:j+1]
u_exact = u_ref[:, 0, j:j+1]
k_exact = k[:, j]
```

```python
t_f_train.shape, f_train.shape, t_u_train.shape, u_train.shape
```

```python
t_f_train.shape
```

```python
t_u_train.shape
```

```python
model = models.PINN(eps=0.0)
```

```python
loss = model.train(t_f_train, f_train, t_u_train, u_train, niter=30000)
# model.log_k = tf.math.log(45.858)
model.restore()
```

```python
tf.math.exp(model.log_k)
```

```python
k_exact
```

```python
u_pred = model.call(
    tf.constant(t, tf.float32)
)

plt.plot(t, u_pred, '--')
plt.plot(t, u_exact)
plt.plot(t_u_train, u_train, 'o')
plt.figure()
plt.plot(t, model.pde(tf.constant(t, tf.float32)), '--')
plt.plot(t, f[:, j])
plt.plot(t_f_train, f_train, 'o')
```
