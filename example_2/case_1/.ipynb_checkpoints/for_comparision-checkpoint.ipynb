```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
GAN = sio.loadmat("GAN.mat")
flows_10 = sio.loadmat("flow_10.mat")
flows_25 = sio.loadmat("flow_25.mat")
flows_50 = sio.loadmat("flow_50.mat")

```

```python
eigs_ref.shape
```

```python
eigs_ref = flows_10["eigs_ref"]
plt.figure(dpi=200)
idx = np.arange(0, 15).astype(np.int32)
plt.plot(idx, eigs_ref[0, :15], "o-", color="black", label="reference")
plt.errorbar(idx, flows_10["mu_realnvp"][0, :15], flows_10["std_realnvp"][0, :15], fmt="s-", color="blue", label="RealNVP")
plt.errorbar(idx, flows_10["mu_maf"][0, :15], flows_10["std_maf"][0,:15], fmt="^-", color="red", label="MAF")
plt.errorbar(idx, flows_10["mu_iaf"][0, :15], flows_10["std_iaf"][0, :15], fmt="+-", color="green", label="IAF")
plt.errorbar(idx, GAN["mu_10"][0, :15], GAN["std_10"][0, :15], fmt="*-", color="purple", label="PI-GAN")
plt.legend()
plt.title("$l=0.1$")
plt.ylabel("Eigen-values")
plt.savefig("compare_10.png")
```
