```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
import tensorflow as tf
```

```python
x = np.linspace(0, 1, 101)
y = np.linspace(0, 1, 101)
xx, yy = np.meshgrid(x, y)
xi = -1 + 2 * np.random.uniform(size=[10, 10000])

xx = xx.reshape([-1, 1])
yy = yy.reshape([-1, 1])
inputs = np.concatenate([xx, yy], axis=-1)

xi = tf.constant(xi, tf.float32)
inputs = tf.constant(inputs, tf.float32)
```

```python
u = xi[0:1, :] * tf.math.sin(1*np.pi*inputs[:, 0:1]) * tf.math.sin(1*np.pi*inputs[:, 1:2]) / (1*np.pi)**2 + \
    xi[1:2, :] * tf.math.sin(2*np.pi*inputs[:, 0:1]) * tf.math.sin(2*np.pi*inputs[:, 1:2]) / (2*np.pi)**2 + \
    xi[2:3, :] * tf.math.sin(3*np.pi*inputs[:, 0:1]) * tf.math.sin(3*np.pi*inputs[:, 1:2]) / (3*np.pi)**2 + \
    xi[3:4, :] * tf.math.sin(4*np.pi*inputs[:, 0:1]) * tf.math.sin(4*np.pi*inputs[:, 1:2]) / (4*np.pi)**2 + \
    xi[4:5, :] * tf.math.sin(5*np.pi*inputs[:, 0:1]) * tf.math.sin(5*np.pi*inputs[:, 1:2]) / (5*np.pi)**2 + \
    xi[5:6, :] * tf.math.sin(6*np.pi*inputs[:, 0:1]) * tf.math.sin(6*np.pi*inputs[:, 1:2]) / (6*np.pi)**2 + \
    xi[6:7, :] * tf.math.sin(7*np.pi*inputs[:, 0:1]) * tf.math.sin(7*np.pi*inputs[:, 1:2]) / (7*np.pi)**2 + \
    xi[7:8, :] * tf.math.sin(8*np.pi*inputs[:, 0:1]) * tf.math.sin(8*np.pi*inputs[:, 1:2]) / (8*np.pi)**2 + \
    xi[8:9, :] * tf.math.sin(9*np.pi*inputs[:, 0:1]) * tf.math.sin(9*np.pi*inputs[:, 1:2]) / (9*np.pi)**2 + \
    xi[9:10, :] * tf.math.sin(10*np.pi*inputs[:, 0:1]) * tf.math.sin(10*np.pi*inputs[:, 1:2]) / (10*np.pi)**2


u_xx = xi[0:1, :] * tf.math.sin(1*np.pi*inputs[:, 0:1]) * tf.math.sin(1*np.pi*inputs[:, 1:2]) + \
       xi[1:2, :] * tf.math.sin(2*np.pi*inputs[:, 0:1]) * tf.math.sin(2*np.pi*inputs[:, 1:2]) + \
       xi[2:3, :] * tf.math.sin(3*np.pi*inputs[:, 0:1]) * tf.math.sin(3*np.pi*inputs[:, 1:2]) + \
       xi[3:4, :] * tf.math.sin(4*np.pi*inputs[:, 0:1]) * tf.math.sin(4*np.pi*inputs[:, 1:2]) + \
       xi[4:5, :] * tf.math.sin(5*np.pi*inputs[:, 0:1]) * tf.math.sin(5*np.pi*inputs[:, 1:2]) + \
       xi[5:6, :] * tf.math.sin(6*np.pi*inputs[:, 0:1]) * tf.math.sin(6*np.pi*inputs[:, 1:2]) + \
       xi[6:7, :] * tf.math.sin(7*np.pi*inputs[:, 0:1]) * tf.math.sin(7*np.pi*inputs[:, 1:2]) + \
       xi[7:8, :] * tf.math.sin(8*np.pi*inputs[:, 0:1]) * tf.math.sin(8*np.pi*inputs[:, 1:2]) + \
       xi[8:9, :] * tf.math.sin(9*np.pi*inputs[:, 0:1]) * tf.math.sin(9*np.pi*inputs[:, 1:2]) + \
       xi[9:10, :] * tf.math.sin(10*np.pi*inputs[:, 0:1]) * tf.math.sin(10*np.pi*inputs[:, 1:2])

# u_xx = xi[0:1, :] * tf.math.sin(np.pi*inputs[:, 0:1]) * tf.math.sin(np.pi*inputs[:, 1:2]) + \
#     xi[1:2, :] * tf.math.sin(2*np.pi*inputs[:, 0:1]) * tf.math.sin(2*np.pi*inputs[:, 1:2]) + \
#     xi[2:3, :] * tf.math.sin(3*np.pi*inputs[:, 0:1]) * tf.math.sin(3*np.pi*inputs[:, 1:2]) + \
#     xi[3:4, :] * tf.math.sin(4*np.pi*inputs[:, 0:1]) * tf.math.sin(4*np.pi*inputs[:, 1:2]) + \
#     xi[4:5, :] * tf.math.sin(5*np.pi*inputs[:, 0:1]) * tf.math.sin(5*np.pi*inputs[:, 1:2]) + \
#     1 * xi[5:6, :] * tf.math.sin(6*np.pi*inputs[:, 0:1]) * tf.math.sin(6*np.pi*inputs[:, 1:2])
u_xx = -u_xx
u_yy = u_xx

f = 0.1 * (u_xx + u_yy) + u * (u ** 2 - 1)
```

```python
u = u.numpy()
f = f.numpy()
u = u.reshape([101, 101, 10000])
f = f.reshape([101, 101, 10000])
```
