```python
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import tensorflow as tf
import time

import models
import flows


data = sio.loadmat("./dataset/data_8.mat")
t = data["t"]
u_ref = data["sols"]
f = data["f"]
k = data["k"]
```

#### Randomly select data points

```python
l = 9999
u_exact = u_ref[:, 0, l:l+1]
f_exact = f[:, l:l+1]
k_exact = k[:, l:l+1]

t_f_train = t[::16]
f_train = f_exact[::16]

np.random.seed(666)
idx = np.random.choice(f_train.shape[0], f_train.shape[0], replace=False)[:5]
f_train = f_train[idx]
t_f_train = t_f_train[idx]

np.random.seed(726991)
idx_u = np.random.choice(u_exact.shape[0], u_exact.shape[0], replace=False)[:10]
# idx_u = [20, 50, 100, 150, 240]
_t_u_train = t[idx_u, :]
_u_train = u_exact[idx_u, :]
```

#### Case 1: One data point

```python
# np.random.seed(3333)
noise_f = 0.0
noise_u = 0.00
u_train = _u_train[:1]
t_u_train = _t_u_train[:1]
f_train = f_train + noise_f * np.random.normal(size=f_train.shape)
u_train = u_train + noise_u * np.random.normal(size=u_train.shape)
print(u_train.shape)
```

```python
#### Load meta and nf
meta = models.MHPINN(num_tasks=2000, dim=50, name="meta_kl_8")
meta.restore()

permutation = list(np.arange(26, 52, 1)) + list(np.arange(0, 26, 1))
nf = flows.MAF(
    dim=52,
    permutation=permutation,
    hidden_layers=[100, 100],
    num_bijectors=10,
    activation=tf.nn.relu,
    name="maf_kl_8",
)
nf.restore()
model = models.Model(body=meta, flow=nf, dim=50, eps=1e-6)
```

```python
loss = model.train(t_f_train, f_train, t_u_train, u_train, niter=30000)
```

```python
model.restore()
u_pred = model.call(tf.constant(t, tf.float32))
f_pred = model.pde(tf.constant(t, tf.float32))
print(tf.math.exp(model.log_k).numpy().flatten(), k_exact)

u_pred = u_pred.numpy()
np.linalg.norm(u_pred - u_exact, 2) / np.linalg.norm(u_exact, 2)
```

```python
fig = plt.figure(dpi=100)
ax = fig.add_subplot()
ax.scatter(t_f_train, f_train, s=5, color="blue", marker="o", facecolor="none", linewidths=8, label="measurements")
ax.plot(t, f_exact, color="black", linestyle="-", linewidth=2, label="reference")
ax.plot(t, f_pred, color="red", linestyle="--", linewidth=2, label="prediction")
ax.set_ylabel("$f$")
ax.set_xlabel("$t$")
plt.savefig("f.png")

fig = plt.figure(dpi=100)
ax = fig.add_subplot()
ax.scatter(t_u_train, u_train, s=5, color="blue", marker="o", facecolor="none", linewidths=8, label="measurements")
ax.plot(t, u_exact, color="black", linestyle="-", linewidth=2, label="reference")
ax.plot(t, u_pred, color="red", linestyle="--", linewidth=2, label="prediction")
ax.set_ylabel("$u$")
ax.set_xlabel("$t$")
plt.savefig("u.png")
```

```python
sio.savemat(
    "gd_1.mat",
    {
        "t": t,
        "u_ref": u_exact,
        "f_ref": f_exact,
        "k_ref": k_exact,
        "f_pred": f_pred.numpy(),
        "u_pred": u_pred,
        "t_u_train": t_u_train, "u_train": u_train,
        "t_f_train": t_f_train, "f_train": f_train,
    }
)
```
