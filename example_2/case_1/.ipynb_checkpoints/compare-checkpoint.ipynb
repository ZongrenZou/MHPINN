```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_probability as tfp
import time
import argparse

import models
import flows
```

```python
def load(l, idx, flow_name):
    file_name = "./data/data_" + str(int(100*l)) + ".mat"
    data = sio.loadmat(file_name)
    t = data["t"]
    u_ref = data["sols"]
    f = data["f"]
    l = data["l"]

    t_train = t[::8]
    f_train = f[::8, :2000]
    u_ref = u_ref[..., :2000]

    model_name = "meta_" + str(int(100 * l)) + "_" + str(int(idx))
    meta = models.Meta(num_tasks=2000, dim=50, name=model_name)
    meta.restore()

    t_test = np.linspace(0, 1, 257).reshape([-1, 1])
    u_pred = meta.call(
        tf.constant(t_test, tf.float32), meta.heads
    ).numpy()

    L2 = np.sqrt(np.sum((u_pred - u_ref[:, 0, :])**2, axis=0) / np.sum(u_ref[:, 0, :]**2, axis=0))
    print(np.mean(L2))

    permutation = list(np.arange(26, 51, 1)) + list(np.arange(0, 26, 1))

    model_name = flow_name + "_" + str(int(100*l)) + "_" + str(int(idx))
    
    if flow_name == "realnvp":
        nf = flows.RealNVP(
            dim=51,
            num_masked=26,
            permutation=permutation,
            hidden_layers=[100, 100],
            num_bijectors=10,
            activation=tf.nn.relu,
            name=model_name,
            eps=0.1,
        )
    elif flow_name == "maf":
        nf = flows.MAF(
            dim=51,
            permutation=permutation,
            hidden_layers=[100, 100],
            num_bijectors=10,
            activation=tf.nn.relu,
            name=model_name,
            eps=0,
        )
    elif flow_name == "iaf":
        nf = flows.IAF(
            dim=51,
            permutation=permutation,
            hidden_layers=[100, 100],
            num_bijectors=10,
            activation=tf.nn.relu,
            name=model_name,
            eps=0,
        )
    # nf.restore()
    nf.load_weights("./results_flows/"+flow_name+"_" + str(int(100*l)) + "/" + model_name)
    
    heads = nf.sample(10000)
    t_test = tf.constant(np.linspace(0, 1, 257).reshape([-1, 1]), tf.float32)
    f_pred = meta.pde(t_test, tf.transpose(heads))
    
    K = tfp.stats.covariance(tf.transpose(f_pred)).numpy() / 256
    eigs = np.real(np.linalg.eigvals(K)).astype(np.float32)
    eigs = eigs[np.argsort(-eigs)]

    K = tfp.stats.covariance(tf.transpose(f)).numpy() /256
    eigs2 = np.real(np.linalg.eigvals(K)).astype(np.float32)
    eigs2 = eigs2[np.argsort(-eigs2)]
    # return meta, nf, eigs, eigs2
    return eigs, eigs2
```

```python
l = 0.1

flow_name = "realnvp"
eigs = []
for idx in range(10):
    eigs += [load(l, idx, flow_name)[0]]
eigs_ref = load(l, 0, flow_name)[1]
eigs = np.stack(eigs, axis=-1)
mu_realnvp = np.mean(eigs, axis=-1)
std_realnvp = np.std(eigs, axis=-1)

flow_name = "iaf"
eigs = []
for idx in range(10):
    eigs += [load(l, idx, flow_name)[0]]
eigs = np.stack(eigs, axis=-1)
mu_iaf = np.mean(eigs, axis=-1)
std_iaf = np.std(eigs, axis=-1)

flow_name = "maf"
eigs = []
for idx in range(10):
    eigs += [load(l, idx, flow_name)[0]]
eigs = np.stack(eigs, axis=-1)
mu_maf = np.mean(eigs, axis=-1)
std_maf = np.std(eigs, axis=-1)


sio.savamat(
    "flow_10.mat",
    {
        "mu_realnvp": mu_realnvp,
        "std_realnvp": std_realnvp,
        "mu_maf": mu_maf, "std_maf": std_maf,
        "mu_iaf": mu_iaf, "std_iaf": std_iaf,
        "eigs_ref": eigs_ref,
    }
)
```

```python
plt.figure(dpi=200)
idx = np.arange(0, 15).astype(np.int32)
plt.plot(idx, eigs_ref[:15], "o-", color="black", label="reference")
plt.errorbar(idx, mu_realnvp[:15], std_realnvp[:15], fmt="s-", color="blue", label="RealNVP")
plt.errorbar(idx, mu_maf[:15], std_maf[:15], fmt="^-", color="red", label="MAF")
plt.errorbar(idx, mu_iaf[:15], std_iaf[:15], fmt="+-", color="green", label="IAF")
plt.legend()
plt.title("$l=0.1$")
plt.ylabel("Eigen-values")

plt.show()
sio.savemat(
    "flow_10.mat",
    {
        "mu_realnvp": mu_realnvp,
        "std_realnvp": std_realnvp,
        "mu_maf": mu_maf, "std_maf": std_maf,
        "mu_iaf": mu_iaf, "std_iaf": std_iaf,
        "eigs_ref": eigs_ref,
    }
)
```
