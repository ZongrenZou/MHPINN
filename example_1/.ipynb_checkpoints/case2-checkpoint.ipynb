#### Distribution learning

$$\theta = f(\xi) = \xi \frac{\sigma}{s} + \mu \longrightarrow \xi = f^{-1}(\theta) = s\frac{\theta - \mu}{\sigma}
\\\longrightarrow p_\theta(\theta) = p_\xi(f^{-1}(\theta))\text{det}\frac{df^{-1}}{d\theta} = p_\xi(f^{-1}(\theta))\prod_{i=1}^N\frac{s_i}{\sigma_i}.$$

Hence, we have $$\log p_\theta(\theta) = \log p_\xi(f^{-1}(\theta)) + \sum_{i=1}^N\log\frac{s}{\sigma_i} = \log p_\xi(f^{-1}(\theta)) + N\log s - \sum_{i=1}^N\log \sigma_i.$$

```python
permutation = list(np.arange(26, 51, 1)) + list(np.arange(0, 26, 1))

nf = flows.MAF(
    dim=51,
    permutation=permutation,
    hidden_layers=[100, 100],
    num_bijectors=10,
    activation=tf.nn.relu,
    name="maf",
)
nf.restore()
```

```python
heads = nf.sample(1000)
x_test = tf.constant(np.linspace(-1, 1, 100).reshape([-1, 1]), tf.float32)
# y_test = tf.constant(y_train, tf.float32)

out = meta.call(x_test, tf.transpose(heads))
for i in range(1000):
    plt.plot(x_test, out[:, i])
plt.ylim([-6, 6])

plt.figure()
out = meta.call(x_test, meta.heads)
for i in range(1000):
    plt.plot(x_test, out[:, i])
plt.ylim([-6, 6])
plt.show()
```

#### Meta-learning, with UQ

```python
import neuraluq as neuq
```

```python
# np.random.seed(1234) # for case 1
np.random.seed(6666) # for case 2
# p.random.seed(666)
noise = 0.2
x_test = np.linspace(-1, 1, 100).reshape([-1, 1])
# idx = np.random.choice(50, 50, replace=False)[:5]
# x_train = x_test[idx, :]

# # case 1
# x_train = np.linspace(-0.9, -0.1, 4).reshape([-1, 1])
# A = 2
# w = 2 * np.pi
# k = +1

# case 2
x_train = np.linspace(-0.9, -0.1, 5).reshape([-1, 1])
A = 2
w = 4 * np.pi
k = -1

y_train = A * np.cos(w * x_train) + 2 * k * x_train
y_test = A * np.cos(w * x_test) + 2 * k * x_test
y_train += noise * np.random.normal(size=y_train.shape)
```

```python
plt.plot(x_train, y_train, 'o')
plt.plot(x_test, y_test)
```

```python
class Samplable(neuq.variables._Samplable):
    
    def __init__(self, nf):
        super().__init__()
        self._num_tensors = 1
        self._initial_values = [nf.sample([])]
        self.log_prob_fn = tf.function(nf.log_prob)
    
    def log_prob(self, samples):
        return self.log_prob_fn(samples[0])
    

class Surrogate(neuq.surrogates.Surrogate):

    def __init__(self, meta_model):
        self.meta_model = meta_model
        
    def __call__(self, inputs, var_list):
        # xi = self.flow.std / self.flow.scale * var_list[0] + self.flow.mu
        xi = tf.reshape(var_list[0], [-1, 51])
        return self.meta_model.call(inputs, tf.transpose(xi))
```

```python
prior = Samplable(nf)
surrogate = Surrogate(meta)
process = neuq.process.Process(surrogate=surrogate, prior=prior)

likelihood = neuq.likelihoods.Normal(
    inputs=x_train,
    targets=y_train, 
    sigma=noise, 
    processes=[process],
)

model = neuq.models.Model(
    processes=[process], likelihoods=[likelihood],
)
```

```python
method = neuq.inferences.HMC(
    num_samples=1000,
    num_burnin=1000,
    init_time_step=0.1,
    leapfrog_step=30,
    seed=6666,
)
model.compile(method)
```

```python
t0 = time.time()
samples, results = model.run()
t1 = time.time()
print("Acceptance rate: %.3f \n"%(np.mean(results)))  # if HMC is used
print("Elapsed of HMC: ", t1 - t0)
```

```python
heads = tf.transpose(samples[0])
y_pred = meta.call(
    tf.constant(x_test, tf.float32), heads 
)
y_pred = y_pred.numpy()

mu = np.mean(y_pred, axis=1)
std = np.std(y_pred, axis=1)
# std = np.sqrt(std**2 + noise**2)
```

```python
fig = plt.figure(dpi=200)
ax = fig.add_subplot()
ax.scatter(x_train, y_train, s=5, color="blue", marker="o", facecolor="none", linewidths=8, label="measurements")
ax.plot(x_test, y_test, color="black", linestyle="-", linewidth=2, label="reference")
ax.plot(x_test, mu, color="red", linestyle="--", linewidth=2, label="mean")
ax.fill_between(x_test.flatten(), (mu+2*std), (mu-2*std), alpha=0.3, color="green", label="2 std")
# ax.legend(loc=0)
ax.set_ylim([-5, 5])
ax.set_box_aspect(2/3)
ax.set_title("HMC")
# fig.savefig("case_2_2_hmc.png")
```

#### Laplace approximation

```python
# # np.random.seed(1234) # for case 1
# np.random.seed(6666) # for case 2
# # p.random.seed(666)
# noise = 0.05
# x_test = np.linspace(-1, 1, 100).reshape([-1, 1])
# idx = np.random.choice(100, 100, replace=False)[:5]
# x_train = x_test[idx, :]
# A = 2
# w = 4 * np.pi
# k = -1
# np.random.normal(size=y_train.shape)
# y_train = A * np.cos(w * x_train) + 2 * k * x_train
# y_test = A * np.cos(w * x_test) + 2 * k * x_test
# y_train += noise * np.random.normal(size=y_train.shape)
```

```python
LA = models.LA(body=meta, flow=nf, dim=50, noise=noise)
```

```python
loss = LA.train(x_train, y_train, niter=20000)
```

```python
# LA.restore()
x_train_tf = tf.constant(x_train, tf.float32)
y_train_tf = tf.constant(y_train, tf.float32)

@tf.function
def Hessian():
    with tf.GradientTape() as tape_xx:
        with tf.GradientTape() as tape_x:
            neg_MAP = LA.neg_log_posterior(x_train_tf, y_train_tf)
        neg_MAP_x = tape_x.gradient(neg_MAP, LA.head)# [..., 0]
    neg_MAP_xx = tape_xx.jacobian(neg_MAP_x, LA.head)# [..., 0]
    return neg_MAP_xx
```

```python
H = Hessian()
cov = tf.linalg.inv(H[:, 0, :, 0]).numpy()
cov = (cov + cov.T) / 2 # + 1e-8 * np.eye(51)
mean = LA.head.numpy().flatten()
```

```python
samples = np.random.multivariate_normal(mean, cov, size=1000)
samples = tf.constant(samples, tf.float32)
```

```python
# heads = tf.transpose(samples * LA.flow.std / LA.flow.scale + LA.flow.mu)
heads = tf.transpose(samples)
y_pred = meta.call(
    tf.constant(x_test, tf.float32), heads
)
y_pred = y_pred.numpy()

mu = np.mean(y_pred, axis=1)
std = np.std(y_pred, axis=1)
# std = np.sqrt(std**2 + noise**2)
```

```python
fig = plt.figure(dpi=200)
ax = fig.add_subplot()
ax.scatter(x_train, y_train, s=5, color="blue", marker="o", facecolor="none", linewidths=8, label="measurements")
ax.plot(x_test, y_test, color="black", linestyle="-", linewidth=2, label="reference")
ax.plot(x_test, mu, color="red", linestyle="--", linewidth=2, label="mean")
ax.fill_between(x_test.flatten(), (mu+2*std), (mu-2*std), alpha=0.3, color="green", label="2 std")
# ax.legend(loc=0)
ax.set_ylim([-5, 5])
ax.set_box_aspect(2/3)
ax.set_title("LA")
# fig.savefig("case_2_2_la.png")
# fig.savefig("case_1_3_no_regularized.png")
# plt.savefig("case_1_3_no_regularized.png")
# plt.savefig("case_1_3_regularized.png")
```
