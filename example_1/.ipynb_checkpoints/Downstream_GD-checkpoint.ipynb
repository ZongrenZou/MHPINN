```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import tensorflow as tf

import models
import flows
```

```python
mhnn = models.MHNN(num_tasks=1000, dim=50, name="mhnn")
mhnn.restore()

permutation = list(np.arange(26, 51, 1)) + list(np.arange(0, 26, 1))
nf = flows.MAF(
    dim=51,
    permutation=permutation,
    hidden_layers=[100, 100],
    num_bijectors=10,
    activation=tf.nn.relu,
    name="maf",
)
nf.restore()
```

#### Figure 1

```python
# np.random.seed(1234) # for case 1
np.random.seed(6666) # for case 2
# np.random.seed(666)
x_test = np.linspace(-1, 1, 100).reshape([-1, 1])
# idx = np.random.choice(100, 100, replace=False)[:3]
# x_train = x_test[idx, :]

# case 1
x_train = np.linspace(-0.9, -0.1, 4).reshape([-1, 1])
A = 2
w = 2 * np.pi
k = +1

# # case 2
# x_train = np.linspace(-0.9, -0.1, 5).reshape([-1, 1])
# A = 2
# w = 4 * np.pi
# k = -1

# # case 3: OOD
# x_train = np.linspace(-0.9, -0.1, 5).reshape([-1, 1])
# A = 2
# w = 1.5 * np.pi
# k = +1

y_train = A * np.cos(w * x_train) + 2 * k * x_train
y_test = A * np.cos(w * x_test) + 2 * k * x_test

plt.plot(x_train, y_train, 'o')
plt.plot(x_test, y_test)
```

```python
model = models.Downstream(mhnn=mhnn, flow=nf, dim=50, eps=1e-3)
```

```python
loss = model.train(x_train, y_train, niter=20000)
```

```python
model.restore()
y_pred = model.call(tf.constant(x_test, tf.float32))

fig = plt.figure(dpi=200)
ax = fig.add_subplot()
ax.scatter(x_train, y_train, s=5, color="blue", marker="o", facecolor="none", linewidths=8, label="measurements")
ax.plot(x_test, y_test, color="black", linestyle="-", linewidth=2, label="reference")
ax.plot(x_test, y_pred.numpy(), color="red", linestyle="--", linewidth=2, label="prediction")
# ax.legend(loc=0)
ax.set_ylim([-5, 5])
ax.set_xlabel("$x$")
ax.set_ylabel("$u$")
ax.set_box_aspect(2/3)
# fig.savefig("case_1_3_regularized.png")
# fig.savefig("case_1_3_no_regularized.png")
# plt.savefig("case_1_3_no_regularized.png")
# plt.savefig("case_1_3_regularized.png")
```

```python
results = {
    "x_train": x_train, "u_train": y_train,
    "x_test": x_test, "u_test": y_test,
    "u_pred": y_pred.numpy()
}
```

```python
model_ref = models.Reference(mhnn=mhnn, dim=50, eps=0.001)
loss = model_ref.train(x_train, y_train, niter=20000)
```

```python
model_ref.restore()
y_pred = model_ref.call(tf.constant(x_test, tf.float32))

fig = plt.figure(dpi=200)
ax = fig.add_subplot()
ax.scatter(x_train, y_train, s=5, color="blue", marker="o", facecolor="none", linewidths=8, label="measurements")
ax.plot(x_test, y_test, color="black", linestyle="-", linewidth=2, label="reference")
ax.plot(x_test, y_pred.numpy(), color="red", linestyle="--", linewidth=2, label="prediction")
# ax.legend(loc=0)
ax.set_ylim([-5, 5])
ax.set_xlabel("$x$")
ax.set_ylabel("$u$")
ax.set_box_aspect(2/3)
# fig.savefig("case_1_3_regularized.png")
# fig.savefig("case_1_3_no_regularized.png")
# plt.savefig("case_1_3_no_regularized.png")
# plt.savefig("case_1_3_regularized.png")
```

```python
results.update(
    {"u_pred_ref_1": y_pred.numpy()}
)
```

```python
model_ref2 = models.NN(eps=0.001)
loss = model_ref2.train(x_train, y_train, niter=20000)
```

```python
model_ref2.restore()
y_pred = model_ref2.call(tf.constant(x_test, tf.float32))

fig = plt.figure(dpi=200)
ax = fig.add_subplot()
ax.scatter(x_train, y_train, s=5, color="blue", marker="o", facecolor="none", linewidths=8, label="measurements")
ax.plot(x_test, y_test, color="black", linestyle="-", linewidth=2, label="reference")
ax.plot(x_test, y_pred.numpy(), color="red", linestyle="--", linewidth=2, label="prediction")
# ax.legend(loc=0)
ax.set_ylim([-5, 5])
ax.set_xlabel("$x$")
ax.set_ylabel("$u$")
ax.set_box_aspect(2/3)
# fig.savefig("case_1_3_regularized.png")
# fig.savefig("case_1_3_no_regularized.png")
# plt.savefig("case_1_3_no_regularized.png")
# plt.savefig("case_1_3_regularized.png")
```

```python
results.update(
    {"u_pred_ref_2": y_pred.numpy()}
)
sio.savemat(
    "case_1.mat", results
)
```
