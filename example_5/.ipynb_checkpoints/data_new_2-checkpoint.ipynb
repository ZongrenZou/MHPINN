```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
# import tensorflow as tf
```

```python
x = np.linspace(0, 2*np.pi, 101)
y = np.linspace(0, 2*np.pi, 101)
xx, yy = np.meshgrid(x, y)

N = 10000
D = 30
s = 4
# xi = -1 + 2 * np.random.uniform(size=[N, D*s])
xi = np.random.uniform(size=[N, D*s])
# xi = np.random.normal(size=[N, D*4])
f = np.zeros([N, 101, 101])

for i in range(D):
    print(s*i, s*i+1, s*i+2, s*i+3)
    f += xi[:, s*i+0:s*i+1][..., None] * np.sin((i+1)*(xx-np.pi)/2/np.pi)[None, ...] * np.sin((i+1)*(yy-np.pi)/2/np.pi)[None, ...] + \
        + xi[:, s*i+1:s*i+2][..., None] * np.cos((i+1)*(xx-np.pi)/2/np.pi)[None, ...] * np.sin((i+1)*(yy-np.pi)/2/np.pi)[None, ...] + \
        + xi[:, s*i+2:s*i+3][..., None] * np.sin((i+1)*(xx-np.pi)/2/np.pi)[None, ...] * np.cos((i+1)*(yy-np.pi)/2/np.pi)[None, ...] + \
        + xi[:, s*i+3:s*i+4][..., None] * np.cos((i+1)*(xx-np.pi)/2/np.pi)[None, ...] * np.cos((i+1)*(yy-np.pi)/2/np.pi)[None, ...]

# for i in range(D):
#     print(s*i, s*i+1, s*i+2, s*i+3)
#     f += xi[:, s*i+0:s*i+1][..., None] * np.sin((i+1)*(xx)/2/np.pi)[None, ...] * np.sin((i+1)*(yy)/2/np.pi)[None, ...] + \
#         + xi[:, s*i+1:s*i+2][..., None] * np.cos((i+1)*(xx)/2/np.pi)[None, ...] * np.sin((i+1)*(yy)/2/np.pi)[None, ...] + \
#         + xi[:, s*i+2:s*i+3][..., None] * np.sin((i+1)*(xx)/2/np.pi)[None, ...] * np.cos((i+1)*(yy)/2/np.pi)[None, ...] + \
#         + xi[:, s*i+3:s*i+4][..., None] * np.cos((i+1)*(xx)/2/np.pi)[None, ...] * np.cos((i+1)*(yy)/2/np.pi)[None, ...]



f = 1/D * f 
```

```python
sio.savemat(
    "source_120_2.mat",
    {
        "f": f, "x": x, "y": y, "xx": xx, "yy": yy,
        "xi": xi
    }
)
```
