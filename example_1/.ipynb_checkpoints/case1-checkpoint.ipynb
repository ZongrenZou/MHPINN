## Case 1: no distribution learning

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import tensorflow as tf

import models
```

```python
data = sio.loadmat("case1.mat")
x1 = data["x1"]
x2 = data["x2"]
y1 = data["y1"]
y2 = data["y2"]
w1 = data["w1"]
w2 = data["w2"]
A = data["A"]
k = data["k"]

x1_train = x1[0, :].reshape([-1, 1])
y1_train = y1[:1000].T
x2_train = x2[0, :].reshape([-1, 1])
y2_train = y2[:1000].T
w1_train = w1[:1000]
w2_train = w2[:1000]
A1_train = A[:1000]
A2_train = A[2000:3000]
k1_train = k[:1000]
k2_train = k[2000:3000]
```

```python
meta = models.Meta(num_tasks=1000, dim=50)
```

```python
# loss = meta.train(
#     x1_train,
#     y1_train,
#     x2_train,
#     y2_train,
#     niter=50000,
#     ftol=1e-6,
# )
```

```python
meta.restore()
# loss[-1]
```

#### To see if training is sufficient

```python
x_test = np.linspace(-1, 1, 100).reshape([-1, 1])
out1 = meta.call(
    tf.constant(x_test, tf.float32), meta.heads[:, :500]
)

i = 3
plt.plot(x_test, out1[:, i], '--')
plt.plot(x1_train, y1_train[:, i], '+')
plt.plot(x_test, A1_train[i] * np.cos(w1_train[i] * x_test) + 1.5 * k1_train[i] * x_test)
```

```python
x_test = np.linspace(-1, 1, 100).reshape([-1, 1])
out2 = meta.call(
    tf.constant(x_test, tf.float32), meta.heads[:, 1000:]
)

i = 100
plt.plot(x_test, out2[:, i], '--')
plt.plot(x2_train, y2_train[:, i], 'o')
plt.plot(x_test, A2_train[i] * np.cos(w2_train[i] * x_test) + 1.5 * k2_train[i] * x_test)
```

#### meta learning

```python
np.random.seed(1234)
x_test = np.linspace(-1, 1, 100).reshape([-1, 1])
idx = np.random.choice(100, 100, replace=False)[:15]
x_train = x_test[idx, :]
A = 2
w = 2 * np.pi
k = 1
y_train = A * np.cos(w * x_train) + 1.5 * k * x_train
y_test = A * np.cos(w * x_test) + 1.5 * k * x_test
```

```python
plt.plot(x_train, y_train, 'o')
plt.plot(x_test, y_test)
```

```python
model = models.Model(body=meta, dim=50, eps=0.0001, scale=1.0)
# model.head = tf.Variable(tf.random.normal(shape=[model.dim+1, 1]), dtype=tf.float32)
loss = model.train(x_train, y_train, niter=40000)
```

```python
model.restore()
```

```python
y_pred = model.call(tf.constant(x_test, tf.float32))

plt.plot(x_train, y_train, 'o')
plt.plot(x_test, y_test)
plt.plot(x_test, y_pred.numpy(), '--')
```
